```python
df['genres'] = df['genres'].fillna('')
```

```python
# converting text data to feature vectore
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfid = TfidfVectorizer()

feature_vector= tfid.fit_transform(df['genres'])
```

```python
print(feature_vector)
```

# cosine similarity 

```python
# getting similarity score using cosine similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(feature_vector)
similarity
```

```python
similarity.shape
```

# creating a list with all movie name as per given in dataset

```python
list_of_all_movie = df['title'].tolist()
list_of_all_movie
```

## getting movie namae from the user

```python
movie_name = input('enter your favourite movie name: ')
```


## finding close match of the movie 

```python

import difflib

find_close_match = difflib.get_close_matches(movie_name, list_of_all_movie)

find_close_match
```

```python
close_match = find_close_match[0]
close_match
```

## finding the index of movie with title

```python
index_of_movie = df[df.title==close_match]['index'].values[0]
index_of_movie
```

## getting list of similar movie

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
print(len(similarity_score))
```

## sorting the movie based on similarity score

```python
sort1 = sorted(similarity_score, key= lambda x:x[1],reverse =True)
sort1
```

## print the name of similar movie based on the movie id 

```python
print('movie suggested for you : \n')

i = 1
for movie in sort1:
    index = movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    
    if (i<=10):
        print ( i , '.', title_from_index)
        
    i+= 1    
```

# movie recommendation system

```python
import difflib
movie_name = input('enter your favourite movie name : ')

list_of_all_movie = df['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_movie)

close_match = find_close_match[0]

index_of_movie = df[df.title==close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

abc = sorted(similarity_score, key= lambda x:x[1],reverse =True)

print('\n movies suggested for you : \n')

print('-----------------')

i = 1
for movie in abc:
    index = movie[0]
    title = df[df.index==index]['title'].values[0]
    
    if (i<=10):
        print(i , '.', title)
        
    i+= 1    
```
